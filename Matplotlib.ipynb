```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
unrate = pd.read_excel('uer-20150702-tyejjd8j.xlsx')
#unrate['月份'] = pd.to_datetime(unrate['月份'])
#print(unrate)
first_twelve = unrate[0:12]
print(first_twelve)
plt.plot(first_twelve['月份'],first_twelve['失业率(%)'])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Unemplotment rate")
plt.title("The USA's Unemplotment rate")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(4,3,1)
ax2 = fig.add_subplot(4,3,3)
ax3 = fig.add_subplot(4,3,5)
plt.show()
```

```python
fig = plt.figure(figsize = (6,6))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))
plt.show()
```

```python
#根据不同颜色来画折线图
fig = plt.figure(figsize=(10,6))
colors =['red','blue','green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 +i)
    plt.plot(subset['月份'],subset['失业率(%)'],c = colors[i],label=label)

plt.legend(loc='best')
plt.show()
```
