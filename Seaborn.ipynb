```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-1)*flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

# 调色板
- color_palette()能传入任何Matplotlib所支持的颜色
- color_palette()不写参数则默认颜色
- set_palette()设置所有图的颜色

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

# 圆形画板
最常用的方法是使用hls的颜色空间，这是RGB值得一个简单转换

```python
sns.palplot(sns.color_palette("hls",12))
```

```python
data = np.random.normal(size = (20,8))+np.arange(8)/2
sns.boxplot(data=data,palette = sns.color_palette("hls",8))
```

hls_palette()函数来控制颜色的亮度和饱和度
- l-亮度 lightness
- s-饱和度 saturation

```python
sns.palplot(sns.hls_palette(8,l=.5,s=.9))
```

```python
sns.palplot(sns.color_palette("Paired",10))
```

# 使用xkcd颜色来命名颜色
xkcd包含了一套众包努力的针对随机RGB色的命名，产生了954个可以随时通过xdcd_rgb字典种调用的命名颜色

```python
plt.plot([0,1],[0,1],sns.xkcd_rgb["pale red"],lw =3)
plt.plot([0,1],[0,2],sns.xkcd_rgb["medium green"],lw =3)
plt.plot([0,1],[0,3],sns.xkcd_rgb["denim blue"],lw =3)
```

# 连续色板
色彩随数据变换，比如数据越来越重要，则颜色越来越深

```python
sns.palplot(sns.color_palette("Blues"))
```

```python
# 如果想要颜色进行翻转渐变，可以在面板名称后面添加一个_r后缀
sns.palplot(sns.color_palette("Blues_r"))
sns.palplot(sns.color_palette("BuGn_r"))
```
